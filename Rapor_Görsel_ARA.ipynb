```python
import numpy as np
import matplotlib.pyplot as plt
import os
from random import randint
```

```python
# Hubble clean - dis - model 2çift
# TUG ham - model çift
#
```

```python
#Hubble ara
dis_dir = "E:\\YAD\\YAD_İşlenmiş_Düzgün\\Test\\"
model_dir = "D:\\pred\\"

dis_paths= os.listdir("E:\\YAD\\YAD_İşlenmiş_Düzgün\\")
model_paths = os.listdir("D:\\pred\\")
```

```python
idx = randint(0,18208)
```

```python
pred = np.load(os.path.join(model_dir, model_paths[idx]))
sample = np.load(os.path.join(dis_dir, model_paths[idx][len("predmap"):]))


dis = sample[0].reshape(512,512)
clean = sample[1].reshape(512,512)
pred = pred[0].reshape(512,512) + dis
pred[pred<0] = 0
clean_dis = np.abs(clean - dis)
clean_pred = np.abs(clean - pred)

# Create a figure and axes
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Set vmin and vmax for consistent contrast
vmin = 0.0
vmax = 0.2

# Plot clean image
axes[0, 0].imshow(clean[384:512, 384:512], cmap="gray", vmin=vmin, vmax=vmax)
axes[0, 0].set_title("Temiz Hubble Görüntüsü", weight="bold")
axes[0, 0].axis('off')

# Plot distorted image
axes[0, 1].imshow(dis[384:512, 384:512], cmap="gray", vmin=vmin, vmax=vmax)
axes[0, 1].set_title("Bozulmuş Hubble Görüntüsü", weight="bold")
axes[0, 1].axis('off')

# Plot predicted image
axes[0, 2].imshow(pred[384:512, 384:512], cmap="gray", vmin=vmin, vmax=vmax)
axes[0, 2].set_title("Model ile Düzeltilmiş Hubble Görüntüsü", weight="bold")
axes[0, 2].axis('off')

# Plot clean-distorted residue map
axes[1, 1].imshow(clean_dis[384:512, 384:512], cmap="gray", vmin=0, vmax=vmax)
axes[1, 1].set_title("Temiz - Bozulmuş Kalıntı Haritası", weight="bold")
axes[1, 1].axis('off')

# Plot clean-predicted residue map
axes[1, 2].imshow(clean_pred[384:512, 384:512], cmap="gray", vmin=0, vmax=vmax)
axes[1, 2].set_title("Temiz - Model Tahmini Kalıntı Haritası", weight="bold")
axes[1, 2].axis('off')

# Remove the empty plot (axes[1, 0])
fig.delaxes(axes[1, 0])

# Adjust layout and display the plot
plt.tight_layout()

plt.savefig("C:\\Users\\hp\\Desktop\\Hubble14.png", dpi = 300)

plt.show()
```
