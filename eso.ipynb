```python
import matplotlib.pyplot as plt

from astropy.wcs import WCS
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename

filename = get_pkg_data_filename('C:\\Users\\hp\\Desktop\\04A\\TUG\\915\\Image\\2017-05-03\\T100_915_DATE_2017-05-03_TIME_23.42.30.99_light.fit')

hdu = fits.open(filename)[0]
wcs = WCS(hdu.header)

plt.figure(figsize=(30,30))
plt.subplot(projection=wcs)
plt.imshow(hdu.data[1000:1512, 1000:1512], vmin=-0, vmax=1000, origin='lower')
plt.grid(color='gray', ls='solid')
plt.xlabel('Galactic Longitude')
plt.ylabel('Galactic Latitude')
```
