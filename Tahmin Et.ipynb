```python
#LOSS
def loss(y_true, y_pred):
    error = tf.abs(y_true - y_pred)
    weights = tf.abs(y_true)
    #weights = (tf.abs(y_true) + tf.abs(y_pred)) / 2
    weighted_loss = weights*error
    return tf.reduce_mean(weighted_loss)
#LOSS
```

```python
model = tf.keras.models.load_model("C:\\Users\\kullanıcı\\Desktop\\YAD_04A\\Model6.keras", compile = False)

adam = tf.keras.optimizers.Adam(learning_rate = 0.001)
model.compile(optimizer=adam, loss = loss, metrics=[loss, "mse", "mae"])
```

```python
dire = "E:\\YAD\\YAD_İşlenmiş_Düzgün\\Test\\"
pred_dir = "H:\\pred\\"
```

```python
for filename in os.listdir(dire):
    if filename.endswith(".npy"):
        if os.path.isfile(pred_dir + "predmap" + filename):
            continue
        
        initial = np.load(os.path.join(dire, filename))

        pred = model.predict(initial[0].reshape(1, 512, 512, 1))
        np.save(pred_dir+"predmap"+filename, pred)
```
